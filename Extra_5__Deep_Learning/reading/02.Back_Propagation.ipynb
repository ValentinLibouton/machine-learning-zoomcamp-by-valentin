# Back Propagation

Un résumé rapide de ce qu'est un réseau neuronal :

Un réseau neuronal est un groupe d'unités d'E/S connectées où chaque connexion a un poids associé à ses programmes informatiques. Il permet de construire des modèles prédictifs à partir de grandes bases de données. Ce modèle s'inspire du système nerveux humain. Il vous aide à comprendre les images, l'apprentissage humain, la parole informatique, etc.

## What is Backpropagation?

La rétropropagation est l'essence même de la formation des réseaux neuronaux. Il s'agit de la méthode de réglage fin des poids d'un réseau neuronal sur la base du taux d'erreur obtenu à l'époque précédente (c'est-à-dire à l'itération). Un réglage correct des poids permet de réduire les taux d'erreur et de rendre le modèle fiable en augmentant sa généralisation.

La rétropropagation est une forme abrégée de "rétropropagation des erreurs". Il s'agit d'une méthode standard de formation des réseaux neuronaux artificiels. Cette méthode permet de calculer le gradient d'une fonction de perte par rapport à tous les poids du réseau.

## How Backpropagation Works: Simple Algorithm 
Consider the following diagram :

![back propagation](./img/back.png)

1. Les entrées X, arrivent par le chemin préconnecté
2. Les entrées sont modélisées à l'aide de poids réels W. Les poids sont généralement choisis au hasard.
3. Calculer la sortie pour chaque neurone depuis la couche d'entrée jusqu'à la couche de sortie, en passant par les couches cachées.
4. Calculer l'erreur dans les sorties
5. Revenir de la couche de sortie à la couche cachée pour ajuster les poids de manière à réduire l'erreur.

## Why We Need Backpropagation?

Les principaux avantages de la rétropropagation sont les suivants :

- La rétropropagation est rapide, simple et facile à programmer
- Il n'y a pas de paramètres à régler en dehors du nombre d'entrées.
- C'est une méthode flexible car elle ne nécessite pas de connaissances préalables sur le réseau.
- Il s'agit d'une méthode standard qui fonctionne généralement bien.
- Elle ne nécessite aucune mention spéciale des caractéristiques de la fonction à apprendre.

## What is a Feed Forward Network?
Un réseau neuronal feedforward est un réseau neuronal artificiel dont les nœuds ne forment jamais de cycle. Ce type de réseau neuronal comporte une couche d'entrée, des couches cachées et une couche de sortie. Il s'agit du premier type de réseau neuronal artificiel et du plus simple.

## Types of Backpropagation Networks
Two Types of Backpropagation Networks are:

- Static Back-propagation
- Recurrent Backpropagation

### Static back-propagation:
Il s'agit d'un type de réseau de rétropropagation qui produit une correspondance entre une entrée statique et une sortie statique. Il est utile pour résoudre les problèmes de classification statique tels que la reconnaissance optique de caractères.

### Recurrent Backpropagation:
La rétropropagation récurrente est alimentée vers l'avant jusqu'à ce qu'une valeur fixe soit atteinte. Ensuite, l'erreur est calculée et propagée vers l'arrière.

La principale différence entre ces deux méthodes est que le mappage est rapide dans la rétropropagation statique alors qu'il est non statique dans la rétropropagation récurrente.

### Backpropagation Key Points 

- Simplifie la structure du réseau en intégrant des liens pondérés qui ont le moins d'effet sur le réseau formé.
- Vous devez étudier un groupe de valeurs d'entrée et d'activation pour développer la relation entre les couches d'entrée et d'unité cachée.
- Cela permet d'évaluer l'impact d'une variable d'entrée donnée sur la sortie d'un réseau. Les connaissances acquises grâce à cette analyse doivent être représentées sous forme de règles.
- La rétropropagation est particulièrement utile pour les réseaux neuronaux profonds travaillant sur des projets sujets aux erreurs, tels que la reconnaissance d'images ou de la parole.
- La rétropropagation tire parti des règles de chaîne et de puissance, ce qui lui permet de fonctionner avec n'importe quel nombre de sorties.

### Disadvantages of using Backpropagation
- Les performances réelles de la rétropropagation sur un problème spécifique dépendent des données d'entrée.
- La rétropropagation peut être très sensible aux données bruitées.
- Vous devez utiliser l'approche basée sur la matrice pour la rétropropagation au lieu du mini-batch.
